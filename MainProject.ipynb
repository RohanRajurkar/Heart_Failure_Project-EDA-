```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
#Seperating/Slicing Columns on x,y axis
x = df.iloc[:, 1:] #Independent Data 2D
y = df.iloc[:, 0] #Dependent 1D-series
x.dtypes
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder()
x[["sex","ejection_fraction"]] = oe.fit_transform(x[["sex","ejection_fraction"]])
#Fit-tranform are seperate functions
#fit is to perform Calculations on data
#Tranform is to apply that calculations
```

```python
#Training & Testing
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest =train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
from sklearn.metrics import mean_squared_error

# Assuming you already have your regression model fitted and predictions made
ypred = model.predict(xtest)

# Calculate Mean Squared Error
mse = mean_squared_error(ytest, ypred)

print("Mean Squared Error:", mse)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
from sklearn.metrics import r2_score
print(f"Accuracy : {r2_score(ytest, ypred)}")
```
